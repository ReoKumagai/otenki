```python

# 地域名と対応するURLを保持したcsvファイルを作成する
  
import requests
from bs4 import BeautifulSoup

import csv
from google.colab import files

import re

url = 'https://www.data.jma.go.jp/obd/stats/etrn/select/prefecture00.php?prec_no=14&block_no=47412&year=1990&month=01&day=&view='

r = requests.get(url)

soup = BeautifulSoup(r.content, 'lxml')

str1 = 'https://www.data.jma.go.jp/obd/stats/etrn/select/'

for area in soup.find_all('area'):
  
    pref = area.get('href')
    
    prefname = area.get('alt')
    
    haruka = str1+pref
    
    r2 = requests.get(haruka)
    
    soup2 = BeautifulSoup(r2.content, 'lxml')
    
    haru = []
    
    haru2 =[]
    
    haru3 =[]
    
    for area in soup2.find_all('area'):
      
      aga = area.get('alt')
      
      agh = area.get('href')
      
      if  '県' in aga:
        pass
      elif '東京都' in aga:
        pass
      elif '大阪府' in aga:
        pass
      elif '京都府' in aga:
        pass
      elif '地方' in aga:
        pass
      else:
        haru.append(aga)
        
        
        url2 = agh
        m1 = re.search('prec_no=[0-9]+', url2)
        haru2.append(m1.group())
              
        m2 = re.search('block_no=[0-9]+', url2)
        haru3.append(m2.group())
        
        
        
    
    nharu = haru[::2]
    
    nharu2 = haru2[::2]
    
    nharu3 = haru3[::2]
    
    areanames = nharu #The list of area name

    precs = nharu2 #The list of prec number
    
    blocks = nharu3 #The list of block number
    
    slist = list(zip(areanames, precs, blocks)) #integrated list

    with open(prefname + '.csv', 'w', encoding = 'utf-16') as filewriting:
        writer = csv.writer(filewriting, dialect = 'excel-tab', lineterminator = '\n')
        writer.writerows(slist)

    files.download(prefname + '.csv')   
              
```
